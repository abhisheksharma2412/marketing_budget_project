```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (update the path if needed)
df = pd.read_csv("marketing_campaign_dataset.csv")

# Display the first few rows and summary statistics
print("First 5 rows of the dataset:")
print(df.head())
print("\nSummary Statistics:")
print(df.describe())

# --- Exploratory Data Analysis (EDA) ---
# Plotting the distribution of key numeric features
numeric_cols = ['ROI', 'Clicks', 'Impressions', 'Acquisition_Cost']

plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(8,6))
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix of Key Features")
plt.show()

# --- Feature Engineering ---
# Ensure Acquisition_Cost is numeric: remove $ sign or commas if needed
df['Acquisition_Cost'] = df['Acquisition_Cost'].replace('[\$,]', '', regex=True).astype(float)

# 1. Ratio Features
# Clicks per Impression
df['Clicks_per_Impression'] = df['Clicks'] / df['Impressions']

# Cost per Click
df['Cost_per_Click'] = df['Acquisition_Cost'] / df['Clicks']

# 2. Temporal Features: Convert 'Date' to datetime and extract month, day, quarter
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Quarter'] = df['Date'].dt.quarter

# 3. Interaction Feature: Combine Campaign_Type and Channel_Used
# This creates a new categorical feature which might capture the interaction effect
df['Campaign_Channel'] = df['Campaign_Type'].astype(str) + "_" + df['Channel_Used'].astype(str)

# Show head of new features
print("\nDataset with New Features:")
print(df[['Clicks', 'Impressions', 'Clicks_per_Impression', 'Acquisition_Cost', 'Cost_per_Click', 'Date', 'Month', 'Day', 'Quarter', 'Campaign_Channel']].head())

# Visualize new ratio features
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['Clicks_per_Impression'], kde=True, bins=30)
plt.title("Distribution of Clicks per Impression")

plt.subplot(1, 2, 2)
sns.histplot(df['Cost_per_Click'], kde=True, bins=30)
plt.title("Distribution of Cost per Click")
plt.tight_layout()
plt.show()

# Optionally, visualize the interaction feature counts
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Campaign_Channel', order=df['Campaign_Channel'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Count of Campaign_Channel Categories")
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score

# Load the dataset
df = pd.read_csv("marketing_campaign_dataset.csv")

# Convert Duration column to numeric
df['Duration'] = df['Duration'].str.replace(' days', '', regex=False).astype(int)

# Ensure Acquisition_Cost is numeric (if not already)
df['Acquisition_Cost'] = df['Acquisition_Cost'].replace('[\$,]', '', regex=True).astype(float)

# Perform one-hot encoding for categorical columns (as an example)
categorical_cols = ['Company', 'Campaign_Type', 'Target_Audience', 'Channel_Used', 'Location', 'Language', 'Customer_Segment']
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Define features and target variable
target_col = 'ROI'
features = [col for col in df_encoded.columns if col not in ['ROI', 'Campaign_ID', 'Date']]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(df_encoded[features], df_encoded[target_col], test_size=0.2, random_state=42)

# Train the Random Forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Make predictions and evaluate the model
y_pred = rf.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Random Forest MAE: {mae:.3f}")
print(f"Random Forest R²: {r2:.3f}")
```

```python
import pandas as pd
import numpy as np

# Load your dataset (update the file path if necessary)
df = pd.read_csv("marketing_campaign_dataset.csv")

# ------------------------------
# Preprocessing Steps
# ------------------------------

# Convert 'Duration' (e.g., "30 days") to numeric
df['Duration'] = df['Duration'].str.replace(' days', '', regex=False).astype(int)

# Clean 'Acquisition_Cost' column by removing '$' and commas, then convert to float
df['Acquisition_Cost'] = df['Acquisition_Cost'].replace('[\$,]', '', regex=True).astype(float)

# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# ------------------------------
# Derived Feature Engineering
# ------------------------------

# Create derived ratio features if they don't already exist
if 'Clicks_per_Impression' not in df.columns:
    df['Clicks_per_Impression'] = df['Clicks'] / df['Impressions']

if 'Cost_per_Click' not in df.columns:
    df['Cost_per_Click'] = df['Acquisition_Cost'] / df['Clicks']

# Combine Target_Audience and Customer_Segment to create an interaction feature
df['Audience_Segment'] = df['Target_Audience'].astype(str) + '_' + df['Customer_Segment'].astype(str)

# Extract detailed temporal features from 'Date'
df['Day_of_Week'] = df['Date'].dt.day_name()         # e.g., Monday, Tuesday, etc.
df['Is_Weekend'] = df['Day_of_Week'].isin(['Saturday', 'Sunday']).astype(int)
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Quarter'] = df['Date'].dt.quarter

# Optional: Create a holiday indicator using US Federal Holidays
try:
    from pandas.tseries.holiday import USFederalHolidayCalendar
    cal = USFederalHolidayCalendar()
    holidays = cal.holidays(start=df['Date'].min(), end=df['Date'].max())
    df['Is_Holiday'] = df['Date'].isin(holidays).astype(int)
except ImportError:
    print("USFederalHolidayCalendar not available. Skipping holiday indicator.")

# ------------------------------
# Transformation: Check and Transform Skewed Features
# ------------------------------

# Define a list of numeric columns to evaluate
numeric_cols = ['Acquisition_Cost', 'Clicks', 'Impressions', 
                'Cost_per_Click', 'Clicks_per_Impression', 'ROI']

# Compute skewness for the numeric features
skew_vals = df[numeric_cols].skew()
print("Skewness of numeric features:")
print(skew_vals)

# Set a threshold for skewness; features with absolute skew above this will be transformed
skew_threshold = 0.75

# Apply log1p transformation (log(1+x)) to handle zeros gracefully
for col in numeric_cols:
    if abs(skew_vals[col]) > skew_threshold:
        new_col = col + '_log'
        df[new_col] = np.log1p(df[col])
        print(f"Applied log1p transformation to '{col}', created '{new_col}'.")

# ------------------------------
# Display Updated DataFrame
# ------------------------------

print("\nDataset with Derived Features and Transformations (first 5 rows):")
print(df.head())
```
